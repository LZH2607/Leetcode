```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import urllib
import time
import numpy as np
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
import math
```

```python
# 打开浏览器
driver = webdriver.Chrome(r"./chromedriver.exe")
```

```python
problem_table = pd.read_excel("Problems.xlsx")
problem_table
```

```python
tag_table = pd.read_excel("Tags.xlsx")
tag_table
```

```python
tag = tag_table["标签"]
url = tag_table["URL"]
page = tag_table["页数"]
```

```python
tag_url_dic = {}
```

```python
tag_tag = tag[0]
tag_tag
```

```python
tag_url = url[0]
tag_url
```

```python
tag_page = page[0]
tag_page
```

```python
js = "window.open('" + tag_url + "')"
driver.execute_script(js)
driver.switch_to.window (driver.window_handles[-1])
```

```python
tag_problem_url = []
```

```python
for i in range(1, tag_page+1):
    print(i)
    time.sleep(5)
    n = len(driver.find_elements_by_css_selector("div > section > div > div > div > div > div > ul > li > a"))
    for k in range(1, n+3):
        if driver.find_element_by_css_selector("div > section > div > div > div > div > div > ul > li:nth-child("+str(k)+")").text == str(i):
            driver.find_element_by_css_selector("div > section > div > div > div > div > div > ul > li:nth-child("+str(k)+")").click()
            break
    time.sleep(5)
    
    tag_a_list = driver.find_elements_by_css_selector("a.e1ceh2223")
    print(len(tag_a_list))
    for i in range(0, len(tag_a_list)):
        print(tag_a_list[i].get_attribute("href"))
        tag_problem_url.append(tag_a_list[i].get_attribute("href"))
    
    print(len(tag_problem_url))
```

```python
tag_url_dic[tag_tag] = tag_problem_url
```

```python
for t in range(1, len(tag)):
    tag_tag = tag[t]
    tag_url = url[t]
    tag_page = page[t]
    print(tag_tag, tag_page, tag_url)
    
    js = "window.open('" + tag_url + "')"
    driver.execute_script(js)
    driver.switch_to.window (driver.window_handles[-1])
    
    tag_problem_url = []
    
    if tag_page > 1:
        for i in range(1, tag_page+1):
            print(i)
            time.sleep(5)
            n = len(driver.find_elements_by_css_selector("div > section > div > div > div > div > div > ul > li > a"))
            for k in range(1, n+3):
                if driver.find_element_by_css_selector("div > section > div > div > div > div > div > ul > li:nth-child("+str(k)+")").text == str(i):
                    driver.find_element_by_css_selector("div > section > div > div > div > div > div > ul > li:nth-child("+str(k)+")").click()
                    break
            time.sleep(5)

            tag_a_list = driver.find_elements_by_css_selector("a.e1ceh2223")
            print(len(tag_a_list))
            for i in range(0, len(tag_a_list)):
                # print(tag_a_list[i].get_attribute("href"))
                tag_problem_url.append(tag_a_list[i].get_attribute("href"))

            print(len(tag_problem_url))
    else:
        time.sleep(10)
        tag_a_list = driver.find_elements_by_css_selector("a.e1ceh2223")
        print(len(tag_a_list))
        for i in range(0, len(tag_a_list)):
            # print(tag_a_list[i].get_attribute("href"))
            tag_problem_url.append(tag_a_list[i].get_attribute("href"))

        print(len(tag_problem_url))
    
    tag_url_dic[tag_tag] = tag_problem_url
```
